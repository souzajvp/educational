# Data 

For this example we are using the [CO2 dataset](https://www.statsmodels.org/devel/datasets/generated/co2.html) downloaded from statsmodels

```python
# Loading dataset
data = sm.datasets.co2.load_pandas()
# Setting y as the variable
y = data.data
```

For this dataset, we have 2225 observations of atmospheric CO2 concentrations derived from flask air samples at sites in the SIO network.

```python
y.head()
```

```python
y.shape
```

We can resample the data in buckets by start of the month

```python
# The 'MS' string groups the data in buckets by start of the month
y = y['co2'].resample('MS').mean()
```

We have some data points missing. In this sense, we will apply back fill to the dataset

```python
# The term bfill means that we use the value before filling in missing values
y = y.fillna(y.bfill())
```

Visualizing the time series

```python
y.plot(figsize=(8,4));
```

For the CO2 data we note how there is an important tendency of grow in the CO2 levels. 

## Seasonal decomposition
Time series data can be decomposed into three components:
1. Trend - general behavior of the series;
2. Seasonality - short-cycle effect of the data;
3. Residuals - random noise that can not be explained by the model


The statsmodels package has many utilities for time series analysis. One of them is the [seasonal_decompose](https://www.statsmodels.org/stable/generated/statsmodels.tsa.seasonal.seasonal_decompose.html) method.

```python
from statsmodels.tsa.seasonal import seasonal_decompose
```

```python
decomp = seasonal_decompose(y,model='additive')
```

### Trend

```python
fig, ax = plt.subplots(figsize=(10,4))
decomp.trend.plot(ax=ax, xlabel='')
ax.set_title('Trend component of our TS', y=-0.24,x=0.2, weight='bold', fontsize=16);
```

Analysing the trend data it becomes clear how there is a tendency of growth in the CO2 levels ever since the begining of our data

### Seasonality

```python
fig, ax = plt.subplots(figsize=(15,4))
decomp.seasonal.plot(ax=ax, xlabel='')
ax.set_title('Trend component of our TS', y=-0.24,x=0.12, weight='bold', fontsize=16);
```

In adition to the strong trend in the data we note how there is a clear presence of a seasonal component.

### Residuals

```python
fig, ax = plt.subplots(figsize=(10,4))
decomp.resid.plot(ax=ax, xlabel='')
ax.set_title('Residual component of our TS', y=-0.24,x=0.2, weight='bold', fontsize=16);
```

The residuals reflect the random noise present in the data. Ideally, we want the residuals to be as low as possible so our TS could be *perfectly* represented by the trend and seasonal components.

Every TS behaves differently and can be influenced by a myriad of external factors. The stock price of a company can increase or decrease based on a financial scandal; The ammount of flu cases can increase because of the circulation of a particular strain of the Influenza virus... These events might impact the time series in a form that could not be predicted and lead to a high residual

```python
p = d = q = range(0, 4)

# Generate all different combinations of p, q and q triplets
pdq = list(itertools.product(p, d, q))

# Generate all different combinations of seasonal p, q and q triplets
seasonal_pdq = [(x[0], x[1], x[2], 12) for x in list(itertools.product(p, d, q))]

print('Examples of parameter combinations for Seasonal ARIMA...')
print('SARIMAX: {} x {}'.format(pdq[1], seasonal_pdq[1]))
print('SARIMAX: {} x {}'.format(pdq[1], seasonal_pdq[2]))
print('SARIMAX: {} x {}'.format(pdq[2], seasonal_pdq[3]))
print('SARIMAX: {} x {}'.format(pdq[2], seasonal_pdq[4]))
```

```python
from statsmodels.tsa.arima.model import ARIMA
```

```python
warnings.filterwarnings("ignore") # specify to ignore warning messages

for param in pdq:
    try:
        mod = ARIMA(y,order=param,
                    enforce_stationarity=False,
                    enforce_invertibility=False)

        results = mod.fit()

        print(f'ARIMA{param} - AIC:{results.aic}')
    except:
        continue
```

```python
warnings.filterwarnings("ignore") # specify to ignore warning messages

for param in pdq:
    for param_seasonal in seasonal_pdq:
        try:
            mod = sm.tsa.statespace.SARIMAX(y,
                                            order=param,
                                            seasonal_order=param_seasonal,
                                            enforce_stationarity=False,
                                            enforce_invertibility=False)

            results = mod.fit()

            print('ARIMA{}x{}12 - AIC:{}'.format(param, param_seasonal, results.aic))
        except:
            continue
```

```python
# (0, 0, 1)x(1, 1, 3, 12)12
mod = sm.tsa.statespace.SARIMAX(y,
                                order=(1, 1, 1),
                                seasonal_order=(1, 1, 1, 12),
                                enforce_stationarity=False,
                                enforce_invertibility=False)

results = mod.fit()

print(results.summary().tables[1])
```

```python
results.plot_diagnostics(figsize=(15, 12))
plt.show()
```

```python
pred = results.get_prediction(start=pd.to_datetime('1998-01-01'), dynamic=False)
pred_ci = pred.conf_int()
```

```python
ax = y['1990':].plot(label='observed')
pred.predicted_mean.plot(ax=ax, label='One-step ahead Forecast', alpha=.7)

ax.fill_between(pred_ci.index,
                pred_ci.iloc[:, 0],
                pred_ci.iloc[:, 1], color='k', alpha=.2)

ax.set_xlabel('Date')
ax.set_ylabel('CO2 Levels')
plt.legend()

plt.show()
```

```python
y_forecasted = pred.predicted_mean
y_truth = y['1998-01-01':]

# Compute the mean square error
mse = ((y_forecasted - y_truth) ** 2).mean()
print('The Mean Squared Error of our forecasts is {}'.format(round(mse, 2)))
```

```python
pred_dynamic = results.get_prediction(start=pd.to_datetime('1998-01-01'), dynamic=True, full_results=True)
pred_dynamic_ci = pred_dynamic.conf_int()
```

```python
ax = y['1990':].plot(label='observed', figsize=(20, 15))
pred_dynamic.predicted_mean.plot(label='Dynamic Forecast', ax=ax)

ax.fill_between(pred_dynamic_ci.index,
                pred_dynamic_ci.iloc[:, 0],
                pred_dynamic_ci.iloc[:, 1], color='k', alpha=.25)

ax.fill_betweenx(ax.get_ylim(), pd.to_datetime('1998-01-01'), y.index[-1],
                 alpha=.1, zorder=-1)

ax.set_xlabel('Date')
ax.set_ylabel('CO2 Levels')

plt.legend()
plt.show()
```

```python
# Extract the predicted and true values of our time series
y_forecasted = pred_dynamic.predicted_mean
y_truth = y['1998-01-01':]

# Compute the mean square error
mse = ((y_forecasted - y_truth) ** 2).mean()
print('The Mean Squared Error of our forecasts is {}'.format(round(mse, 2)))
```

```python
pred_uc = results.get_forecast(steps=500)

# Get confidence intervals of forecasts
pred_ci = pred_uc.conf_int()
```

```python
ax = y.plot(label='observed', figsize=(20, 15))
pred_uc.predicted_mean.plot(ax=ax, label='Forecast')
ax.fill_between(pred_ci.index,
                pred_ci.iloc[:, 0],
                pred_ci.iloc[:, 1], color='k', alpha=.25)
ax.set_xlabel('Date')
ax.set_ylabel('CO2 Levels')

plt.legend()
plt.show()
```

```python
srag = pd.read_csv('./data/srag_cases_2010-2020.csv')
srag = srag.query('ANO != 2020')
srag['DT_NOTIFIC'] = pd.to_datetime(srag['DT_NOTIFIC'], format='%Y-%m-%d')
```

```python
y = srag[['DT_NOTIFIC', 'CASOS']].set_index('DT_NOTIFIC')['CASOS']
```

```python
# Define the p, d and q parameters to take any value between 0 and 2
p = d = q = range(0, 2)

# Generate all different combinations of p, q and q triplets
pdq = list(itertools.product(p, d, q))

# Generate all different combinations of seasonal p, q and q triplets
seasonal_pdq = [(x[0], x[1], x[2], 12) for x in list(itertools.product(p, d, q))]

print('Examples of parameter combinations for Seasonal ARIMA...')
print('SARIMAX: {} x {}'.format(pdq[1], seasonal_pdq[1]))
print('SARIMAX: {} x {}'.format(pdq[1], seasonal_pdq[2]))
print('SARIMAX: {} x {}'.format(pdq[2], seasonal_pdq[3]))
print('SARIMAX: {} x {}'.format(pdq[2], seasonal_pdq[4]))
```

```python
warnings.filterwarnings("ignore") # specify to ignore warning messages

lowest = 100000
for param in pdq:
    for param_seasonal in seasonal_pdq:
        try:
            mod = sm.tsa.statespace.SARIMAX(y,
                                            order=param,
                                            seasonal_order=param_seasonal,
                                            enforce_stationarity=False,
                                            enforce_invertibility=False)

            results = mod.fit()
            if results.aic <= lowest:
                lowest = results.aic
            print('ARIMA{}x{}12 - AIC:{}'.format(param, param_seasonal, results.aic))
        except:
            continue
```

```python
mod = sm.tsa.statespace.SARIMAX(y,
                                order=(0, 1, 1),
                                seasonal_order=(0, 1, 1, 12),
                                enforce_stationarity=False,
                                enforce_invertibility=False)

results = mod.fit()

print(results.summary().tables[1])
```

```python
results.plot_diagnostics(figsize=(15, 12))
plt.show()
```

```python
pred = results.get_prediction(start=pd.to_datetime('2019-01-31'), dynamic=False)
pred_ci = pred.conf_int()
```

```python
ax = y['1990':].plot(label='observed')
pred.predicted_mean.plot(ax=ax, label='One-step ahead Forecast', alpha=.7)

ax.fill_between(pred_ci.index,
                pred_ci.iloc[:, 0],
                pred_ci.iloc[:, 1], color='k', alpha=.2)

ax.set_xlabel('Date')
ax.set_ylabel('CO2 Levels')
plt.legend()

plt.show()
```

```python
y_forecasted = pred.predicted_mean
y_truth = y['2019-01-31':]

# Compute the mean square error
mse = ((y_forecasted - y_truth) ** 2).mean()
print('The Mean Squared Error of our forecasts is {}'.format(round(mse, 2)))
```

```python
pred_dynamic = results.get_prediction(start=pd.to_datetime('2019-01-31'), dynamic=True, full_results=True)
pred_dynamic_ci = pred.conf_int()
```

```python
ax = y['2019':].plot(label='observed', figsize=(20, 15))
pred_dynamic.predicted_mean.plot(label='Dynamic Forecast', ax=ax)

ax.fill_between(pred_dynamic_ci.index,
                pred_dynamic_ci.iloc[:, 0],
                pred_dynamic_ci.iloc[:, 1], color='k', alpha=.25)

ax.fill_betweenx(ax.get_ylim(), pd.to_datetime('1998-01-01'), y.index[-1],
                 alpha=.1, zorder=-1)

ax.set_xlabel('Date')
ax.set_ylabel('CO2 Levels')

plt.legend()
plt.show()
```
